```python
from statsmodels.formula.api import ols
```

```python
import pandas as pd
import io
df = pd.read_csv(io.BytesIO(upload['PortfolioDataClean_2020StackOverflowSurveyResults.csv']))
print(df)
```

```python
model = ols('JobSatConverted ~ Age + ConvertedComp + C(Region) + C(EdLevel) + C(Employment) + C(Ethnicity) + C(Gender) + C(Sexuality) + WorkWeekHrs + YearsCodePro', data=df)
fitted_model = model.fit()
fitted_model.summary()
```
